```python
# Read the 'transcript' data
transcript_path = os.path.join(DATASET_PATH, 'transcript.csv')
transcript_df = pd.read_csv(transcript_path)
```

```python
# Convert data in string to dictionary
import ast

transcript_df['value_dict'] = transcript_df['value'].apply(ast.literal_eval)
```

```python
# Check the information of the dataframe
transcript_df.info()
```

```python
# Extract the key and value from 'value_dict' and create new columns
transcript_df["value_key"] = transcript_df.value_dict.apply(lambda x:list(x.keys())[0])
transcript_df["value_value"] = transcript_df.value_dict.apply(lambda x:list(x.values())[0])
```

```python
# View all unique values in 'value_key'
unique_values = transcript_df['value_key'].unique()
print(unique_values)
```

```python
# Filter for data with 'offer_id' and view the first index
offer_id_data = transcript_df[transcript_df['value_dict'].apply(lambda x: 'offer_id' in x)]
print(offer_id_data['value_dict'].iloc[0])
```

```python
# Filter for data with 'offer id' and view the first index
offer_wo_id_data = transcript_df[transcript_df['value_dict'].apply(lambda x: 'offer id' in x)]
print(offer_wo_id_data['value_dict'].iloc[0])
```
